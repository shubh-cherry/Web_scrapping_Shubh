```python
# Importing libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import date
```

```python
url = "https://m.imdb.com/search/name/?birth_monthday=06-20&ref_=nv_cel_brn"
```

```python
# Access page using request module

page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content,"html.parser")
print(soup.prettify())
```

```python
scrapped_names = soup.find_all('h3',class_ = 'lister-item-header')
scrapped_names
```

```python
names = []
for name in scrapped_names:
    name = name.get_text().replace('\n', "")
    name.strip(" ")
    
    names.append(name)
names
```

```python
data = pd.DataFrame()
data['celib Names'] = names
data['Birthdate'] = date.today()
data
```

```python
data.to_csv('_CELEB_BORN_TODAY.csv',index = False)
```
